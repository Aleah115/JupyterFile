```python
cir_name = [1,2,3,4,5,6,7]
cir_list = [[1],[2],[3],[4],[5],[6],[7]]
for _ in range(6):
    temp = []
    for item in cir_list:
        for i in cir_name:
            num = list(set(item+[i]))
            if not (num in cir_list) and not (num in temp):
                temp.append(num)
    cir_list = cir_list+temp
```

```python
len(cir_list)
```

```python
name = ['asset','meanprice','day','outstanding','roa','leverage','top']
```

```python
res = []
for item in cir_list:
    temp=[]
    for n in item:
        temp.append(name[n-1])
    res.append(temp)
```
