```python
def get_diff_df(df_left, df_right):
    
    divi = pd.merge(df_left, df_right, on=['Segment_a', 'Segment_s'], how='outer')
    divi.dropna(how = 'any', inplace = True)
    divi.reset_index(inplace = True)
    divi = divi.drop('index', axis = 1)
    divi['diff_a'] = abs(divi['总资产_y'] - divi['总资产_x'])
    divi['diff_s'] = abs(divi['流通三板股_y'] - divi['流通三板股_x'])
    divi['diff'] = divi['diff_a'] + divi['diff_s']
    
    return divi
```

```python
#导入数据
path = 'C:/Users/84678/Desktop/做市商效率/科创板做市交易.xlsx'
df_tech_c = pd.read_excel(path, sheet_name="科创板竞价")
df_tech_m = pd.read_excel(path, sheet_name="科创板做市")
```

```python
df_tech_list = [df_tech_c, df_tech_m]
```

```python
df_tech_c
```

```python
#删去总资产和流通三板股有空值的数据_new
for f in df_tech_list:
    f.dropna(how = 'any', inplace = True)
    f.drop(0,inplace = True)
    f.reset_index(inplace=True)
    f[['总资产','流通股合计']] = f[['总资产','流通股合计']].apply(pd.to_numeric)
    del f['index']
```

```python
df_tech_m
```

```python
bins = 3
for f in df_tech_list:
    # df.cat.codes把各类型用整数编码代替
    f['Segment_a'] = bins - pd.qcut(f['总资产'], bins).cat.codes
    f['Segment_s'] = bins - pd.qcut(f['流通股合计'], bins).cat.codes
```

```python
divi_tech = pd.merge(df_tech_list[1], df_tech_list[0], on=['Segment_a','Segment_s'], how="outer")
divi_tech.dropna(how="any", inplace=True)
```

```python
def get_diff_tech_df(df_left, df_right):
    
    divi = pd.merge(df_left, df_right, on=['Segment_a', 'Segment_s'], how='outer')
    divi.dropna(how = 'any', inplace = True)
    divi.reset_index(inplace = True)
    divi = divi.drop('index', axis = 1)
    divi['diff_a'] = abs(divi['总资产_y'] - divi['总资产_x'])
    divi['diff_s'] = abs(divi['流通股合计_y'] - divi['流通股合计_x'])
    divi['diff'] = divi['diff_a'] + divi['diff_s']
    
    return divi
```

```python
df_tech_divi = get_diff_tech_df(df_tech_list[1], df_tech_list[0])
```

```python
list(df_tech_divi.groupby(['代码_x']))[0][1]
```

```python
#为每个做市企业找出对应的差值最小的对照企业
temp_tech_x = df_tech_divi.groupby(['代码_x'])['diff'].idxmin().to_frame()
temp_tech = df_tech_divi.iloc[temp_tech_x.iloc[:,0]]
```

```python
#一个对照企业只保留其差值最小的做市企业，其余删去
temp_tech_y = temp_tech.groupby(['代码_y'])['diff'].idxmin().to_frame()
df_tech_con = df_tech_divi.iloc[temp_tech_y.iloc[:,0]]
```
