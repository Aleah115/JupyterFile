```python
# name_list = ["b_c", "b_m", "i_c", "i_m"]
df_list = [df_b_c, df_b_m, df_i_c, df_i_m]
```

```python
#删去总资产和流通三板股有空值的数据_new
for f in df_list:
    f.dropna(how = 'any', inplace = True)
    f.reset_index(inplace=True)
    del f['index']
```

```python
bins = 5
```

```python
for f in df_list:
    # df.cat.codes把各类型用整数编码代替
    f['Segment_a'] = bins - pd.qcut(f['总资产'], bins).cat.codes
    f['Segment_s'] = bins - pd.qcut(f['流通三板股'], bins).cat.codes
```

```python
def get_diff_df(df_left, df_right):
    
    divi = pd.merge(df_left, df_right, on=['Segment_a', 'Segment_s'], how='left')
    divi.dropna(how = 'any', inplace = True)
    divi.reset_index(inplace = True)
    divi = divi.drop('index', axis = 1)
    divi['diff'] = abs(divi['总资产_y'] - divi['总资产_x'])
    
    return divi
```
